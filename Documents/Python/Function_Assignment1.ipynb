```python
def elegible():
    num = input("Your Gender: ")
    num1 = int(input("Your age: "))

    if num == "male" and num1 >= 21:
        print("Eligible")
    elif num == "female" and num1 >= 18:
        print("Eligible")
    else:
        print("Not Eligible")

elegible()
```

```python
def triangle():
    tri1 = int(input("Height = "))
    tri2 = int(input("Breadth = "))

    print("Area Formula: (Height * Breadth) / 2")
    area = (tri1 * tri2) / 2
    print("Area of Triangle:", area)

    tri3 = int(input("Height1 = "))
    tri4 = int(input("Height2 = "))
    tri5 = int(input("Breadth = "))

    print("Perimeter Formula: Height1 + Height2 + Breadth")
    perimeter = tri3 + tri4 + tri5
    print("Perimeter of Triangle:", perimeter)

triangle()
```
