```python
class  multiple():
    
 def oddeven():
    num=int(input("Enter a number:"))
    if(num%2==0):
        print(num," is a even number")
        
    else:
        print(num, "is a odd number")
     
 def elegible():
    num = input("Your Gender: ")
    num1 = int(input("Your age: "))

    if num == "male" and num1 >= 21:
        print("Eligible")
    elif num == "female" and num1 >= 18:
        print("Eligible")
    else:
        print("Not Eligible")
 def percentage():
  sub1=int(input("Subject1="))
  sub2=int(input("Subject2="))
  sub3=int(input("Subject3="))
  sub4=int(input("Subject4="))
  sub5=int(input("Subject5="))
  add=sub1+sub2+sub3+sub4+sub5
  print("Total:",add)
  percent=add/500*100
  print("Percentage:",percent) 
```

```python
class multiples():

 def triangle1():
    tri1 = int(input("Height = "))
    tri2 = int(input("Breadth = "))

    print("Area Formula: (Height * Breadth) / 2")
    area = (tri1 * tri2) / 2
    print("Area of Triangle:", area)

    tri3 = int(input("Height1 = "))
    tri4 = int(input("Height2 = "))
    tri5 = int(input("Breadth = "))

    print("Perimeter Formula: Height1 + Height2 + Breadth")
    perimeter = tri3 + tri4 + tri5
    print("Perimeter of Triangle:", perimeter)
```
